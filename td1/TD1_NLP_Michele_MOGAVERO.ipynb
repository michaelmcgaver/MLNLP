```python
import nltk

nltk.download('punkt')

tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')
text=" Hello everyone. Hope all are fine and doing well. Hope you find the book interesting"
tokenizer.tokenize(text)

french_tokenizer=nltk.data.load('tokenizers/punkt/french.pickle')
```

```python
with open('extrait_ukraine_20min.txt', 'r', encoding='utf-8') as file:
    file_contents1 = file.read()

text_list1 = [file_contents1]
```

```python
with open('extrait_prof_tué_ladepeche.txt', 'r', encoding='utf-8') as file:
    file_contents2 = file.read()

text_list2 = [file_contents2]
```

```python
with open('extrait_ukraine_lemonde.txt', 'r', encoding='utf-8') as file:
    file_contents3 = file.read()

text_list3 = [file_contents3]
```

```python
for article1 in text_list1:
    article1 = article1.replace('\n', '. ')
for article2 in text_list2:
    article2 = article2.replace('\n', '. ')
for article3 in text_list3:
    article3 = article3.replace('\n', '. ')
article1_tokenised_phrases = french_tokenizer.tokenize(article1)
article2_tokenised_phrases = french_tokenizer.tokenize(article2)
article3_tokenised_phrases = french_tokenizer.tokenize(article3)
```

```python
article1_tokenised_phrases
```

```python
from nltk.tokenize import TreebankWordTokenizer
word_tokenizer = TreebankWordTokenizer()
```

```python
article1_words = []

for i in article1_tokenised_phrases:
    a = word_tokenizer.tokenize(i)
    article1_words.append(a)

article1_words
```

Using regular expressions:

```python
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer("[\w]+")
```

```python
with open('extrait_ukraine_20min.txt', 'r', encoding='utf-8') as file:
    file_contents1 = file.read()
with open('extrait_prof_tué_ladepeche.txt', 'r', encoding='utf-8') as file:
    file_contents2 = file.read()
with open('extrait_ukraine_lemonde.txt', 'r', encoding='utf-8') as file:
    file_contents3 = file.read()
```

```python
tokenized_mots1 = []
tokenized_mots2 = []
tokenized_mots3 = []

for i in article1_tokenised_phrases:
    a = tokenizer.tokenize(i.lower())
    tokenized_mots1.append(a)
for j in article2_tokenised_phrases:
    b = tokenizer.tokenize(j.lower())
    tokenized_mots2.append(b)
for k in article3_tokenised_phrases:
    c = tokenizer.tokenize(k.lower())
    tokenized_mots3.append(c)
tokenized_mots2
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stops=set(stopwords.words('french'))
```

```python
tokenized_mots1_stoppe = []
tokenized_mots2_stoppe = []
tokenized_mots3_stoppe = []
for i in tokenized_mots1:
    a = [word for word in i if word not in stops]
    tokenized_mots1_stoppe.append(a)
for i in tokenized_mots2:
    tokenized_mots2_stoppe.append([word for word in i if word not in stops])
for i in tokenized_mots1:
    tokenized_mots3_stoppe.append([word for word in i if word not in stops])


tokenized_mots1_stoppe
```
